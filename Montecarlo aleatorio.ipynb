```python
def rollgame():
    roll=np.random.randint(1,7)
    if roll >= 4:
        return False
    else:
        return True
```

```python
def apostadorsimple(fondos, apuestainicial, napuestas):
    valor = fondos
    apuesta=apuestainicial
    apuestaActual= 0
    while apuestaActual < napuestas:
        if rollgame:
            valor += apuesta
        else: 
            valor-=apuesta
        apuestaActual+=1
        print("Fondos: ", valor)
```

```python
def apostadorsimple(fondos, apuestainicial, napuestas):
    valor = fondos
    apuesta=apuestainicial
    apuestaActual= 0
    wX, wY= [],[]
    while apuestaActual < napuestas:
        if rollgame():
            valor += apuesta
            wX.append(apuestaActual)
            wY.append(valor)
        else: 
            valor-=apuesta
        apuestaActual+=1
        wX.append(apuestaActual)
        wY.append(valor)
      #  print("Fondos: ", valor)
    plt.plot(wX, wY, lw=1)
```

```python
apostadorsimple(10000,100,500)
```

### Crecimiento de poblaciones "Cadena de Markov"

```python
n=2000 #Tama√±o de la poblaci√≥n m√°xima
a=.5/n #razon de nacimientos
b=.5/n
npasos=1000
x=np.zeros(npasos)
x[0]=100 #poblacion inical
for t in range(npasos-1):
    if 0< x[t]< n-1:
        nacimiento=np.random.randn()
        muerte = np.random.randn()<=b*x[t]
        x[t+1]=x[t]+nacimiento*1-muerte*1
    else:
        x[t+1]=x[t]
```

```python
plt.plot(x,c="orange")
plt.show()
```
